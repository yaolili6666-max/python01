数据缺失值处理

```python
import pandas as pd
import numpy as np
s=pd.Series([1,2,np.nan,3,None,pd.NA]) # 缺失值
print(s)

#查看缺失值
# print(s.isnull())

#处理缺失值
#删除缺失值
print(s.dropna())

#替换缺失值
#用指定值填充缺失值
print(s.fillna(0))

#用前一个非缺失值填充缺失值
print(s.ffill())

```

```python
 #数据变形

 #宽表转换为长表
df = pd.DataFrame({'id':[1,2,3],
                   'name':['edy','lily','james'],
                   'Chinese':[41,51,61],
                   'Math':[71,18,91],
                   'English':[99,121,88],
                   'score':[41,51,61],
                   })
print(df)
df_long = pd.melt(df, id_vars=['id','name'],
        var_name='科目',
        value_name='成绩')
print(df_long)

#
```
