学生成绩统计
创建一个包含10名学生数学成绩的Series，成绩范围在50-100之间。计算平均分、最高分、最低分，并找出高于平均分的学生人数。

```python
import pandas as pd
import numpy as np
print(np.random.randint(50,101,10))
s = pd.Series(np.random.randint(50,101,10))
print(s.mean()) # 计算平均分
print(s.max()) # 计算最高分
print(s.min()) # 计算最低分
print(s[s>s.mean()]) # 找出高于平均分的学生
print(len(s[s>s.mean()])) # 计算高于平均分的学生人数
```

温度数据分析
给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序

```python
t = pd.Series([28, 31, 29, 32, 30, 27, 33], 
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print(t[t>30]) # 找出温度超过30度的天数
print(t.mean()) # 计算平均温度
print(t.sort_values(ascending=False)) # 将温度从高到低排序
```

股票价格分析
给定某股票连续10个交易日的收盘价Series：
计算每日收益率（当日收盘价/前日收盘价 - 1）
找出收益率最高和最低的日期
计算波动率（收益率的标准差）

```python
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                  index=pd.date_range('2023-01-01', periods=10))
print(prices)
#计算每日收益率（当日收盘价-前日收盘价）/前日收盘价
returns = (prices - prices.shift(1)) / prices.shift(1)
print(returns)
#找出收益率最高和最低的日期
print(returns.idxmin())
print(returns.idxmax())
#计算波动率（收益率的标准差）
print(returns.std())
```

数据合并与计算
有两个Series分别记录了某产品在两个城市的日销量：
计算两个城市的总日销量
找出哪个城市的销量更高
计算两个城市销量的相关系数

```python
city_a = pd.Series([120, 135, 140, 130, 145],
                  index=pd.date_range('2023-01-01', periods=5))
city_b = pd.Series([110, 125, 150, 140, 130],
                  index=pd.date_range('2023-01-01', periods=5))

print(city_a + city_b)
print('A' if city_a.sum() > city_b.sum() else 'B')
#计算两个城市销量的相关系数
print(city_a.corr(city_b))
```
