场景：某公司销售数据如下，请完成以下任务：
1.计算每种产品的总销售额（销售额 = 单价 × 销量）。
2.找出销售额最高的产品。
3.按销售额从高到低排序，并输出所有产品信息。

```python
import pandas as pd
import numpy as np
```

```python
data = {
    '产品名称': ['A', 'B', 'C', 'D'],
    '单价': [100, 150, 200, 120],
    '销量': [50, 30, 20, 40]
}
df = pd.DataFrame(data)
print(df)
df['总销售额'] = df[['单价','销量']].prod(axis=1)
print(df)
print(df.nlargest(1,columns='总销售额'))

print(df.sort_values('总销售额',ascending=False))
```

场景：某部门员工考勤数据如下，请完成以下任务：
1.计算每位员工的出勤率（出勤率 = 出勤天数 / 工作日总数）。
2.标记出勤率低于80%的员工。
3.按出勤率从高到低排序。

```python

data = {
    '姓名': ['张三', '李四', '王五', '赵六'],
    '出勤天数': [20, 15, 18, 22],
    '工作日总数': [25, 20, 25, 25]
}
df = pd.DataFrame(data)
df['出勤率'] = df.出勤天数/df.工作日总数
print(df[df.出勤率<0.8])
print(df.sort_values('出勤率',ascending=False))
```

场景：某电影评分数据如下，请完成以下任务：
1.计算每部电影的平均评分。
2.找出评分高于8.5的电影。
3.按平均评分从高到低排序。

```python
data = {
    '电影名称': ['电影A', '电影B', '电影C', '电影D'],
    '评分1': [9.0, 8.5, 8.0, 7.5],
    '评分2': [8.5, 9.0, 8.5, 8.0],
    '评分3': [9.5, 8.0, 7.5, 7.0]
}
df = pd.DataFrame(data)
df['平均评分'] = df[['评分1','评分2','评分3']].mean(axis=1)
print(df['平均评分'])
print(df[df['平均评分']>8.5])
print(df.sort_values('平均评分',ascending=False))
```

场景：某电商平台的用户行为数据如下，请完成以下任务：  
1.计算每位用户的**总消费金额**（消费金额 = 商品单价 × 购买数量）  
2.找出**消费金额最高的用户**，并输出其所有信息  
3.计算所有用户的**平均消费金额**（保留2位小数）  
4.统计**电子产品**的总购买数量  

```python
data = {
    '用户ID': [101, 102, 103, 104, 105],
    '用户名': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    '商品类别': ['电子产品', '服饰', '电子产品', '家居', '服饰'],
    '商品单价': [1200, 300, 800, 150, 200],
    '购买数量': [1, 3, 2, 5, 4]
}
df = pd.DataFrame(data)
df['总消费金额'] = df.商品单价*df.购买数量
print(df)
print(df.sort_values('总消费金额',ascending=False).head(1))

#统计**电子产品**的总购买数量  
print(df[df.商品类别=='电子产品'].购买数量.sum())

```
